```python
import pandas as pd
import numpy as np
import math
```

```python
r=102317272
ar = 0.05 * (r % 7)
br = 0.3 * (r % 5 + 1)
```

```python
df = pd.read_csv("data.csv",encoding="cp1252")
print(df.columns)
x = df["no2"].dropna().values
```

```python
z = []
for val in x:
    z_val = val + ar * math.sin(br * val)
    z.append(z_val)
z = np.array(z)
```

```python
mu = np.mean(z)
var = np.var(z)
lam = 1 / (2 * var)
c = math.sqrt(lam / math.pi)
print("mu =", mu)
print("lambda =", lam)
print("c =", c)
```
